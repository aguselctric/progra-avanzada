```python
class Panelsolar:
    def __init__(self, area_m2, eficiencia):
        self.area = area_m2
        self.eficiencia = eficiencia
        self.activo = True
    def potenciapeak(self):
        if self.activo:
            return self.area * self.eficiencia * 1000
        else:
            return 0
```
