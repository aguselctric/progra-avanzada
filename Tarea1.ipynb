Parte 2. Herencia.
Defina dos subclases, llamadas Guerrero y Mago. Considere que Guerrero tiene un atributo extra llamado furia,
(booleano) que funciona como +10 al ataque. La furia de un guerrero se inicializa en False, pero se activa cuando
su vida es menor o igual al 10% de la vida inicial, luego de recibir el daño. Por otro lado, Mago tiene el atributo
poción, (booleano) que reduce el daño recibido en un factor de 0.8. Un mago activa su poción cuando luego de
recibir el daño, su vida es inferior al 20% de su valor inicial. Activada la furia o la poción, se mantendrán en ese
estado hasta el final de la batalla. Como puede ver, es necesario sobre escribir algunos métodos. 

Parte 3. Composición.
Cree dos clases llamadas Espadas y Hechizos, para los Guerreros y Magos respectivamente. Estas tienen como
atributos el nombre y el factor de daño extra. No olvide usar representaciones legibles. Un Guerrero puede usar
solamente un objeto de la clase Espada y un Mago solamente un objeto de la clase Hechizo. Utilice la función
isinstance de Python. Estos objetos se sortearán antes de cada ataque y deben ser utilizados dentro del método
atacar para modificar el daño.

```python
class Personaje:
    daño_base = 1.2
    total_personajes = 0

    def __init__(self, nombre, vida, nivel, ataque):
        self.nombre = nombre
        self._vida = vida
        self.nivel = nivel
        self.ataque = ataque
        if self._vida <= 0:
            self.estado = "derrotado"
        elif self._vida > 0:   # este solo puede ser "Vivo" o "derrotado"
            self.estado = "Vivo"
        else:
            Exception("Error, el personaje solo puede estar vivo o derrotado")
        Personaje.total_personajes += 1

    def atacar(self, objetivo, arma):
        if self.estado == "derrotado":
            print(f"{self.nombre} no puede atacar porque está derrotado.")
            return
        if objetivo.estado == "derrotado":
            print(f"{objetivo.nombre} ya está derrotado.")
            return
        # Diccionario de armas válidas y su daño
        armas_validas = {
            "espada": 5,
            "arco": 4,
            "hacha": 8,
            "baston": 4
        }
        if arma not in armas_validas:
            print(f"{self.nombre} no puede usar {arma}.")
            return
        # Obtener daño del arma
        daño_arma = armas_validas[arma]

        daño_total = self.ataque * Personaje.daño_base * daño_arma
        # Se completa la acción
        objetivo.recibir_daño(daño_total, self, arma)

        # esto lo dejo comentado porque es de una version anterior que no guarde xd
        #objetivo.recibir_daño(daño_total) # aqui llama a recibir_daño
        #print(f"{self.nombre} atacó a {objetivo.nombre} con {arma} causando {daño_total:.2f} de daño.") #2f para mostrar el daño con dos decimales y no 20
    
    def recibir_daño(self, daño, atacante, arma):
        self._vida -= daño

        print(f"{atacante.nombre} atacó a {self.nombre} con {arma} causando {daño:.2f} de daño.")

        if self._vida <= 0:
            self.estado = "derrotado"
            self._vida = 0
            print(f"{self.nombre} ha sido derrotado.")

    def subir_nivel(self):
        self.nivel += 1
        self.ataque += 5 
        print(f"{self.nombre} subió al nivel {self.nivel}. Ataque aumentado a {self.ataque:.2f}.")

    def esta_vivo(self):
        if self._vida > 0:
            return True
        else:            return False

    def __str__(self):
        return f"{self.nombre} - Vida: {self._vida:.2f}, Nivel: {self.nivel}, Ataque: {self.ataque:.2f}, Estado: {self.estado}"

    @property
    def vida(self):
        return self._vida #a pesar de que se defina como metodo sigue siendo un atributo

    @vida.setter
    def vida(self, valor):
        if valor < 0:
            self._vida = 0
            self.estado = "derrotado"
            print(f"{self.nombre} ha sido derrotado.")
        else:
            self._vida = valor

    def __eq__(self, other):
        if isinstance(other, Personaje): # comprobamos que el otro objeto tambien es una instancia de Personaje
            return self.nivel == other.nivel # comparamos si tienen el mismo nivel
        return False

# hay que añadir una forma de hacer que cada personaje tiene su set de armas y cada arma tiene un daño específico, para que el ataque sea más realista. 
# Por ejemplo, podríamos crear una clase Arma con atributos como nombre y daño, y luego asociar cada personaje con un conjunto de armas que pueda usar en sus ataques.

#Herencia

class Guerrero(Personaje):
    def __init__(self, nombre, vida, nivel, ataque):
        super().__init__(nombre, vida, nivel, ataque)
        self.furia = False
        self.vida_inicial = vida  # Guardamos la vida inicial

    def recibir_daño(self, daño, atacante, arma):
        super().recibir_daño(daño, atacante, arma) # primero recibimos el daño

        # Activar furia si la vida baja al 10% o menos
        if self.vida <= 0.1 * self.vida_inicial:
            self.furia = True

    def atacar(self, objetivo, arma):
        if self.furia:
            self.ataque += 10  # Aumentamos el ataque por furia temporalmente
            super().atacar(objetivo, arma)  # Llamamos al método atacar de antes
            self.ataque -= 10  # Restablecemos el ataque después del ataque con furia
        else:
            super().atacar(objetivo, arma)  # Llamamos al método atacar de antes

# Mago tiene el atributo poción, (booleano) que reduce el daño recibido en un
# factor de 0.8. Un mago activa su poción cuando luego de recibir el daño, su
# vida es inferior al 20% de su valor inicial. Activada la furia o la poción,
# se mantendrán en ese estado hasta el final de la batalla. Como puede ver, es
# necesario sobre escribir algunos métodos.

class Mago(Personaje):
    def __init__(self, nombre, vida, nivel, ataque):
        super().__init__(nombre, vida, nivel, ataque)
        self.vida_inicial = vida
        self.pocion = False

    def recibir_daño(self, daño, atacante, arma):
      if not self.pocion:
        self._vida -= daño
        print(f"{atacante.nombre} atacó a {self.nombre} con {arma} causando {daño:.2f} de daño.")
      if self.pocion:
        daño_final = daño * 0.8
        self._vida -= daño_final
        print(f"{atacante.nombre} atacó a {self.nombre} con {arma} causando {daño_final:.2f} de daño.")
      # Activar poción si baja al 20% (y no se ha usado)
      if not self.pocion and self.vida <= 0.8 * self.vida_inicial:
          self.pocion = True
          print(f"{self.nombre} ha usado una poción de defensa.")
      if self._vida <= 0:
          self.estado = "derrotado"
          self._vida = 0
          print(f"{self.nombre} ha sido derrotado.")

# Cree dos clases llamadas Espadas y Hechizos, para los Guerreros y Magos respectivamente. Estas clases deben tener como
# atributos el nombre y el factor de daño extra. No olvide usar representaciones legibles. Un Guerrero puede usar
# solamente un objeto de la clase Espada y un Mago solamente un objeto de la clase Hechizo. Utilice la función
# isinstance de Python. Estos objetos se sortearán antes de cada ataque y deben ser utilizados dentro del método
# atacar para modificar el daño.

class Espada:
    def __init__(self, nombre, daño_extra):
        self.nombre = nombre
        self.daño_extra = daño_extra

    def __str__(self):
        return f"Espada: {self.nombre}, Daño Extra: {self.daño_extra}"

class Hechizo:
    def __init__(self, nombre, daño_extra):
        self.nombre = nombre
        self.daño_extra = daño_extra

    def __str__(self):
        return f"Hechizo: {self.nombre}, Daño Extra: {self.daño_extra}"

# Crear instancias de espadas y hechizos
espada1 = Espada("Andúril", 2.0)
espada2 = Espada("Glamdring", 1.5)
espada3 = Espada("Dagas de Legolas", 1.8)
espada4 = Espada("Orcrist", 1.2)
hechizo1 = Hechizo("Bola de Fuego", 2.5)
hechizo2 = Hechizo("Rayo", 2.0)
hechizo3 = Hechizo("Escudo Mágico", 1.5)
hechizo4 = Hechizo("You shall not pass", 7.0)


p1 = Personaje("Aragorn", 100, 1, 5)
print(p1)
p2 = Personaje("Legolas", 100, 1, 5)
print(p2)
p3 = Guerrero("Gimli", 120, 1, 4)
print(p3)
p4 = Mago("Gandalf", 150, 1, 4)
print(p4)

# Atacar normalmente (sin furia)
print("\n--- Ataque normal ---")
p3.atacar(p2, "hacha")

# Hacer que Gimli reciba mucho daño para activar furia
print("\n--- Gimli recibe daño para activar furia ---")
p3.recibir_daño(110, p4, "espada")  # lo deja con 10 de vida (≈ 8.3%)

print(f"¿Furia activada? {p3.furia}")
#print(p3) # aqui no muestra el ataque aumentado porque el ataque con furia solo se activa en el método atacar, no en el recibir daño, pero ya se ha activado la furia para el siguiente ataque

# Atacar con furia activada
print("\n--- Ataque con furia ---")
p3.atacar(p2, "hacha")

# Hacer que Gandalf reciba mucho daño para activar poción
print("\n--- Gandalf recibe daño para activar poción ---")
p1.atacar(p4, "espada")
p1.atacar(p4, "espada")
print(p4)
p1.atacar(p4, "espada")
p1.atacar(p4, "espada")
print(p4)
p1.atacar(p4, "espada")
p1.atacar(p4, "espada")
# Activar poción
print(f"¿Poción activada? {p4.pocion}") 
p1.atacar(p4, "baston")
print(p4)
```
